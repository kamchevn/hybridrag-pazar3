```python
import requests
import os
from bs4 import BeautifulSoup
from weasyprint import HTML
def scrape_listing(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.content, "html.parser")

    # Title
    title = soup.find("h1", class_="ci-text-base").get_text(strip=True)

    # Price (new and old) – from value attributes
    new_price_tag = soup.select_one("bdi.new-price span.format-money-int")
    old_price_tag = soup.select_one("bdi.old-price span.format-money-int")
    new_price = new_price_tag["value"] + " ЕУР" if new_price_tag else "непознато"
    old_price = old_price_tag["value"] + " ЕУР" if old_price_tag else None

    # Property details
    tags = {}
    for tag in soup.select(".tags-area a.tag-item"):
        label = tag.find("span").get_text(strip=True).rstrip(":")
        value = tag.find("bdi").get_text(strip=True)
        tags[label] = value

    # Description
    description = soup.select_one(".description-area span").get_text(strip=True)

    # Images
    img_elements = soup.select("div.carousel-inner img")
    images = []
    for img in img_elements:
        src = img.get("src") or img.get("data-src")
        if src and src.startswith("http"):
            images.append(src)

    # Publisher info
    publisher_type = soup.select_one(".company-ad-tag")
    publisher_type = publisher_type.get_text(strip=True) if publisher_type else "Оглас од физичко лице"

    published_date = soup.select_one("bdi.published-date")
    published_time = soup.select_one("bdi.published-time")
    published = f"{published_date.get_text(strip=True)} {published_time.get_text(strip=True)}" if published_date and published_time else "непознато"

    views = soup.select_one("div.views-number span.ci-text-muted")
    views = views.get_text(strip=True) if views else "непознато"

    # Agency info
    agency_name = soup.select_one("h4.ci-text-base span.ci-valign-middle")
    agency_name = agency_name.get_text(strip=True) if agency_name else "непознато"

    agency_address = soup.select_one("p.contact-item span.ci-text-muted")
    agency_address = agency_address.get_text(strip=True) if agency_address else "непозната адреса"

    phone_tag = soup.select_one("p.contact-item a[href^='tel:'] bdi")
    phone = phone_tag.get_text(strip=True) if phone_tag else "непознат број"

    website_tag = soup.select_one("p.contact-item a[href^='http']")
    website = website_tag["href"] if website_tag else "нема веб-страна"

    # Coordinates only (no full address)
    coords_tag = soup.select_one("div.coordinates-block p.ci-text-base")
    if coords_tag:
        raw_coords = coords_tag.get_text(strip=True)
        parts = raw_coords.split()
        if len(parts) == 2:
            coordinates = f"{parts[1]} {parts[0]}"  # lat,lon
        else:
            coordinates = "непознати координати"
    else:
        coordinates = "непознати координати"

    return {
        "title": title,
        "url": url,
        "price": new_price,
        "old_price": old_price,
        "tags": tags,
        "description": description,
        "images": images,
        "publisher_type": publisher_type,
        "published": published,
        "views": views,
        "agency": {
            "name": agency_name,
            "agency_address": agency_address,
            "phone": phone,
            "website": website
        },
        "location": {
            "coordinates": coordinates
        }
    }

def build_html(data):
    html = f"""<html><head><meta charset="utf-8">
    <style>
        body {{ font-family: Arial, sans-serif; margin: 40px; }}
        h1 {{ font-size: 24px; }}
        .url {{ color: blue; font-size: 18px; }}
        .price {{ color: green; font-size: 18px; }}
        .old-price {{ text-decoration: line-through; color: red; }}
        dl {{ margin-top: 20px; }}
        dt {{ font-weight: bold; }}
        dd {{ margin-bottom: 10px; }}
        img {{ max-width: 100%; height: auto; margin: 10px 0; }}
        .meta, .agency, .location {{ font-size: 14px; color: #444; margin-top: 10px; }}
    </style></head><body>
    <h1>{data['title']}</h1>
    <p class='url'>Линк: <a href="{data['url']}">{data['url']}</a></p>
    <p class='price'>Цена: {data['price']}</p>"""

    if data['old_price']:
        html += f"<p class='old-price'>Стара цена: {data['old_price']}</p>"

    html += f"""
    <div class='meta'>
        <p><strong>{data['publisher_type']}</strong></p>
        <p>Објавено: {data['published']}</p>
        <p>Прегледи: {data['views']}</p>
    </div>
    """

    html += "<dl>"
    for label, value in data["tags"].items():
        html += f"<dt>{label}</dt><dd>{value}</dd>"
    html += "</dl>"

    html += f"<p>{data['description']}</p>"

    html += f"""
    <div class="location">
        <h2>Локација на имотот</h2>
        <p>Координати: {data['location']['coordinates']}</p>
        <p><a href="https://www.google.com/maps?q={data['location']['coordinates'].replace(' ', ',')}" target="_blank">Отвори во Google Maps</a></p>
    </div>
    """

    html += f"""
    <div class="agency">
        <h2>Информации за агенцијата</h2>
        <p>Име: {data['agency']['name']}</p>
        <p>Адреса: {data['agency']['agency_address']}</p>
        <p>Телефон: {data['agency']['phone']}</p>
        <p>Веб-страна: <a href="{data['agency']['website']}">{data['agency']['website']}</a></p>
    </div>
    """
    data["images"] = download_images(data["images"])
    for img_url in data["images"]:
        html += f'<div><img src="{img_url}" alt="Слика од куќата"></div>'

    html += "</body></html>"
    return html

def generate_pdf(html, output_filename="listing.pdf"):
    HTML(string=html).write_pdf(output_filename)
    print(f"PDF saved as: {output_filename}")

def download_images(image_urls):
    os.makedirs("images", exist_ok=True)
    local_paths = []
    for i, url in enumerate(image_urls):
        try:
            response = requests.get(url)
            if response.status_code == 200:
                filename = f"images/img_{i}.jpg"
                with open(filename, "wb") as f:
                    f.write(response.content)
                local_paths.append(filename)
        except Exception as e:
            print(f"Failed to download image {url}: {e}")
    return local_paths


base_url = "https://www.pazar3.mk/"
listing_pages = [
    "oglasi/zivealista/stanovi/prodazba",
    "oglasi/zivealista/kukji-vili/prodazba",
    "oglasi/zivealista/stanovi/izdavanje",
    "oglasi/zivealista/kukji-vili/izdavanje"
]

agency_links = []

# Loop over all listing pages
for page in listing_pages:
    page_url = base_url + page
    response = requests.get(page_url)
    response.raise_for_status()
    soup = BeautifulSoup(response.text, "html.parser")

    # Extract only agency ads
    for container in soup.select("div.title.span-col-title"):
        if container.select_one("span.label.label-default.isstore"):
            for a in container.select("a.Link_vis[href]"):
                href = a["href"]
                if href.startswith("//"):
                    href = "https:" + href
                elif href.startswith("/"):
                    href = base_url + href
                agency_links.append(href)

print(f"Found {len(agency_links)} agency ads.")

# Create output folder
os.makedirs("pdf_listings", exist_ok=True)

# Loop through each agency ad
for i, link in enumerate(agency_links, start=1):
    try:
        print(f"Scraping ({i}/{len(agency_links)}): {link}")
        data = scrape_listing(link)
        html = build_html(data)
        pdf_filename = f"pdf_listings/listing_{i}.pdf"
        generate_pdf(html, pdf_filename)
    except Exception as e:
        print(f"Error scraping {link}: {e}")
```
